```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
y=all['NDSI'].values.reshape(-1, 1)
x=all['NDSI_Snow_Cover'].values.reshape(-1, 1)
model.fit(x, y)
a=model.intercept_
b=model.coef_
print('回归系数=',b,'截距=',a)
```

```python
#导入数据拟合函数
from scipy.stats import linregress
import pandas as pd
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from sklearn.metrics import mean_squared_error,r2_score
from matplotlib.pyplot import MultipleLocator

def f_1(x, A, B):  #设置任意拟合曲线
    return A*x + B

if __name__ == '__main__':

    plt.rcParams['font.family'] = ['Arial']
    test_data = pd.read_excel('C:/Users/SZT/Desktop/scatter/pts-all/all-0.xls')
    N = len(test_data)
    x = test_data['NDSI_Snow_Cover'].values.ravel()  # 真实值扁平化
    y = test_data['NDSI'].values.ravel()  # 预测值扁平化
    C = round(r2_score(x, y), 4)#4位相关系数
    rmse = round(np.sqrt(mean_squared_error(x, y)), 3)#3位均方误差

    # 绘制拟合线
    x2 = np.linspace(0, 1)
    y2 = x2
    
    A1, B1 = optimize.curve_fit(f_1, x, y)[0]   #斜率和截距
    y3 = A1 * x + B1
    
    # 开始绘图
    fig, ax = plt.subplots(figsize=(7, 5), dpi=200)
    
       #————————————————————————————————————————————————————————————————————————————
    nbins = 120
    H, xedges, yedges = np.histogram2d(x, y, bins=nbins)
    H = np.rot90(H)
    H = np.flipud(H)
    Hmasked = np.ma.masked_where(H == 0, H)
    plt.pcolormesh(xedges, yedges, Hmasked, cmap=cm.get_cmap('jet'), vmin=0, vmax=10)

    cbar = plt.colorbar(ticks=[0,2,4,6,8,10], drawedges=False)
    
    colorbarfontdict = {'size': 9, "color": 'k'}
    cbar.ax.set_title('Counts', fontdict=colorbarfontdict, pad=8)
    cbar.ax.tick_params(labelsize=10, direction='in')
    cbar.ax.set_yticklabels(['0','4', '8','12', '16','>20' ], family='Times New Roman')
    
#     #————————————————————————————————————————————————————————————————————————————————————
    
    ax.plot(x2, y2, color='k', linewidth=1.5, linestyle='-', zorder=2)#y=x参考线
    ax.plot(x, y3, color='r', linewidth=2, linestyle='-', zorder=2)#数据拟合线

#     # 添加上线和下线
#     up_y2 = 1.15 * x2 + 0.05
#     down_y2 = 0.85*x2 - 0.05

#     ax.plot(x2, up_y2, color='b', lw=1., ls='--', zorder=2, alpha=.8)
#     ax.plot(x2, down_y2, color='b', lw=1., ls='--', zorder=2, alpha=.8)


    #设置横纵坐标
    fontdict1 = {"size": 17, "color": "k", }
    ax.set_xlabel("MODIS NDSI", fontdict=fontdict1)
    ax.set_ylabel("Landsat NDSI ", fontdict=fontdict1)
    
    #设置网格线
#     ax.grid(which='major', axis='y', ls='--', c='k', alpha=.7)
#     ax.set_axisbelow(True)
    #设置边界
    ax.set_xlim((0, 1))
    ax.set_ylim((0, 1))
    
    #设置刻度
#     ax.set_xticks(np.arange(0, 1, step=10))
#     ax.set_yticks(np.arange(0, 1, step=10))
#     # #设置刻度间隔
#     # x_major_locator=MultipleLocator(20)
#     # #把x轴的刻度间隔设置为.5，并存在变量里
#     # y_major_locator=MultipleLocator(20)
#     # ax.xaxis.set_major_locator(x_major_locator)
#     # #把x轴的主刻度设置为.5的倍数
#     # ax.yaxis.set_major_locator(y_major_locator)


    #清除左上右坐标线
    for spine in ['top', 'right']:
        ax.spines[spine].set_visible(None)
#         ax.spines[spine].set_color('k')

#     #排版刻度  =>  漏尖角的
    ax.spines['bottom'].set_color('k')
    ax.tick_params(bottom=True, direction='out', labelsize=14, width=1.5, length=4,left=True)

    
    # 添加题目
    titlefontdict = {"size": 10, "color": "k", }
    ax.set_title('Scatter plot for accuracy verification of MOD10A1 based on Landsat', titlefontdict, pad=20)

    
    #标注误差=>  统计数据
    fontdict = {"size": 12, "color": "k", 'weight': 'bold'}
#     ax.text(0.04, 0.95, r'$R^2=$' + "0.452", fontdict=fontdict) 
#     ax.text(0.04, 0.92, "RMSE=" + str(rmse), fontdict=fontdict)
#     ax.text(0.04, 0.89, r'$y=$' + str(round(A1, 3)) + '$x$' + " + " + str(round(B1, 3)), fontdict=fontdict)
    ax.text(0.04, 0.92, "(c)" , fontdict=fontdict)
#     ax.text(0.04, 0.86, r'$N=$' + str(N), fontdict=fontdict)
    
    plt.savefig('abc.png',width=7,height=5,dpi=900,bbox_inches='tight')
    # ————————————————————————————————————————————————————————————————————————————————————————

    plt.show()
```
